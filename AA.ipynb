```python
plt.scatter(df['budget'], df['c_box_office'], c='blue')
x = [df['budget'].min(), df['budget'].max()]
y = x
plt.plot(x, y, color = 'red')

x = [0, 1, df['budget'].max()]
y = [0, 2, 2*(df['budget'].max())]
plt.plot(x, y, color = 'green')

vmax = max(df['budget'].max(), df['c_box_office'].max())
plt.xlim(0, vmax)
plt.ylim(0, vmax)
```
